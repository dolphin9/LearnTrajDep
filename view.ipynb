```python
import matplotlib.pyplot as plt 
import pandas as pd
```

```python
filename = './checkpoint/logs/main_3dpw_in10_out30_dctn_40.csv'

logs = pd.read_csv(filename)
```

```python
x = logs['epoch']
lr = logs['lr']
loss = logs['t_l']
t_err = logs['t_err']
v_err = logs['v_err']
```

```python
plt.subplot(221)
plt.subplots_adjust(left=1,bottom=2,right=2,top=3)

# equivalent but more general
ax1 = plt.subplot(221)
ax1 = plt.plot(x,lr)
ax1 = plt.ylabel('learn rate')
# add a subplot with no frame
ax2 = plt.subplot(223)
ax2 = plt.plot(x,loss)
ax2 = plt.ylabel('loss')

# add a polar subplot
ax3 = plt.subplot(224)
ax3 = plt.plot(x,t_err)
ax3 = plt.ylabel('t_err')

# add a red subplot that shares the x-axis with ax1
ax4= plt.subplot(222)
ax4 = plt.plot(x,v_err)
ax4 = plt.ylabel('v_err')

plt.show()
```
